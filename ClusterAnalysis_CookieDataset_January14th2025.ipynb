```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Select numerical columns for clustering
numeric_columns = [
    'no of Ingredients',
    'Sweetness Factor (1, too sweet, 10 not sweet enough)',
    'Crunch Factor (1 too little crunch, 10 too much crunch)',
    'Crumble Factor (1, not crumble 10 too much crumble)',
    'Stickiness (0 no, 1 yes)',
    'Appereance (1, is this a cookie? - 10 masterpiece)',
    'moisture (1=dry, 10=moist)',
    'Overall Score (1 to 10)'
]

# Extract numerical data
numerical_data = df[numeric_columns]

# Handle missing values by replacing them with the mean of each column
imputer = SimpleImputer(strategy='mean')
numerical_data_imputed = imputer.fit_transform(numerical_data)

# Standardize the data for K-means
scaler = StandardScaler()
numerical_data_scaled = scaler.fit_transform(numerical_data_imputed)

# Convert back to a DataFrame for better readability
processed_data = pd.DataFrame(numerical_data_scaled, columns=numeric_columns)

# Save processed data to a CSV file
processed_data.to_csv('processed_data.csv', index=False)
print("Processed data saved as 'processed_data.csv'.")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Determine the optimal number of clusters using the Elbow Method and Silhouette Score
inertia = []
silhouette_scores = []
range_clusters = range(2, 11)  # Test for 2 to 10 clusters

for k in range_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(numerical_data_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(numerical_data_scaled, kmeans.labels_))

# Plot the Elbow Method
plt.figure(figsize=(10, 6))
plt.plot(range_clusters, inertia, marker='o')
plt.title("Elbow Method: Optimal Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()

# Plot Silhouette Scores
plt.figure(figsize=(10, 6))
plt.plot(range_clusters, silhouette_scores, marker='o', color='orange')
plt.title("Silhouette Scores for Different Cluster Sizes")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()

# Optimal number can be chosen based on these plots.
```

```python
# Run KMeans clustering with 3 clusters
kmeans_3 = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_3.fit(numerical_data_scaled)

# Add the cluster labels to the original dataset
df['Cluster'] = kmeans_3.labels_

# Display the data with cluster assignments
#tools.display_dataframe_to_user(name="Clustered Data with 3 Clusters", dataframe=df)

# Display cluster centers for interpretation
cluster_centers = pd.DataFrame(kmeans_3.cluster_centers_, columns=numeric_columns)
cluster_centers
```

```python
from sklearn.metrics import pairwise_distances

# Find the coordinates of each centroid using the cluster_centers_ attribute
centroids = kmeans_3.cluster_centers_

# Calculate the euclidean distance between the centroids
centroid_distances = pairwise_distances(centroids)

# Plot distances on heatmap
sns.heatmap(centroid_distances,
            annot=True,
            linewidths=1);
```
